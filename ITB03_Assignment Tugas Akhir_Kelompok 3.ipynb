#### Menghitung total luas panen per komoditas

```python
data = df
data['total_luas_panen'] = df.groupby(['komoditas','tahun'])['luas_panen'].transform('sum')
data
```

#### Menghapus row duplicates

```python
data = data.drop_duplicates(subset=['komoditas','tahun'], ignore_index=True)
data.head(50)
```

#### Menghapus feature yang tidak digunakan

```python
data = data.drop(columns=['kode_kabupaten_kota','luas_panen'])
data
```

#### Reindex feature

```python
data_temp = data.iloc[:,:2]
data = data.iloc [:,2:]
data = pd.concat([data,data_temp], axis=1)
data
```

```python
tes = data.loc[data['komoditas']=='BAWANG MERAH',:]
tes
```

### Membangun model

```python
from sklearn.model_selection import train_test_split
```

```python
X = data['tahun']
y = data['total_luas_panen']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
plt.scatter(X_train, y_train)
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
X_train_ = X_train.values.reshape(-1, 1)
y_train_ = y_train.values.reshape(-1, 1)
regr_model = regr.fit(X_train_, y_train_)

print ('Coefficients: ', regr_model.coef_)
print ('Intercept: ',regr_model.intercept_)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regr_model.coef_[0][0]*X_train + regr_model.intercept_[0], '-r')
plt.show()
```

```python
from sklearn.metrics import r2_score

X_test_ = X_test.values.reshape(-1, 1)
y_test_ = y_test.values.reshape(-1, 1)
y_predict = regr.predict(X_test_)

print("Mean absolute error: %.2f" % np.mean(np.absolute(y_predict - y_test_)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_predict - y_test_ ) ** 2))
print("R2-score: %.2f" % r2_score(y_predict , y_test_) )
```
